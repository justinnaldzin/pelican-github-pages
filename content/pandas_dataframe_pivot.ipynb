```python
# pip3 install pandas openpyxl

import pandas
csv_file = "queries.csv"
queries_dataframe = pandas.read_csv(csv_file)
queries_dataframe.head()
```

```python
pivoted = queries_dataframe.pivot_table(index=['query_id', 'category', 'name'], columns='database', values='query_template', aggfunc=lambda x: ' '.join(x))
#pivoted.columns  # Index(['HANA', 'Oracle Database', 'Oracle Database In-Memory', 'SQL Server'], dtype='object', name='database')
pivoted.index  # MultiIndex(levels=[0, 1, 2],, labels=[0, 1, 2], names=['query_id', 'category', 'name'])
pivoted.to_csv('pivoted.csv')  # flattens header, does NOT include 'database' or 'query_template' names
pivoted.to_excel('pivoted.xlsx')  # flattens header, does NOT include 'database' or 'query_template' names
pivoted.head()
```
